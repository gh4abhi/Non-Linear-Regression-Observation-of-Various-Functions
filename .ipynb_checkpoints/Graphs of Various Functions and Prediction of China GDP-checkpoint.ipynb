```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Linear Function

```python
x=np.arange(-5.0 , 5.0 , 0.1)

y=2*x+3

y_noise=2*np.random.normal(size=x.size)

y_data = y + y_noise

plt.plot(x,y_data,"bo")                                                   #====plt.scatter(x,y_data,color="blue")
plt.plot(x,y,color='red')
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

# Cubic Function

```python
x=np.arange(-5.0 , 5.0 , 0.1)

y=1*(x**3) + 1*(x**2) + 1*x + 3

y_noise=20*np.random.normal(size=x.size)
y_data = y + y_noise

plt.plot(x , y_data , "bo")
plt.plot(x , y , color = "red")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

# Quadratic Function

```python
x=np.arange(-5.0 , 5.0 , 0.1)

y=np.power(x,2)                                # =========   y=x**2
y_noise=2*np.random.normal(size = x.size)
y_data = y + y_noise

plt.plot(x , y_data , "bo")
plt.plot(x , y , color = "red")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

# Exponential Function

```python
x=np.arange(-5.0 , 5.0 , 0.1)

y = np.exp(x)
y_noise = 5*np.random.normal(size=x.size)
y_data = y + y_noise

plt.plot(x , y_data , "bo")
plt.plot(x , y , color = "red")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

# Logarithmic Function

```python
x = np.arange(-5.0 , 5.0 , 0.1)

y=np.log(x)
y_noise = 0.1*np.random.normal(size=x.size)
y_data = y + y_noise

plt.plot(x , y_data , "bo")
plt.plot(x , y , color = "red")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

# Sigmoidal Function

```python
x=np.arange(-5.0 , 5.0 , 0.1)

y=1/(1+np.exp(-1.5*x))
y_noise = 0.1*np.random.normal(size = x.size)
y_data = y +y_noise

plt.plot(x , y_data , "bo")
plt.plot(x , y , color = "red")
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.show()
```

# Non-Linear Regression Project :- Prediction Of China's GDP 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("china_gdp.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df.Year , df.Value , color = "blue")           #       x_data, y_data = (df["Year"].values, df["Value"].values)
plt.xlabel("Year")                                         #       plt.plot(x_data, y_data, 'ro')
plt.ylabel("GDP")
plt.show()                                             
```

# Train/Test Split

```python
msk = np.random.rand(len(df)) < 0.8
train=df[msk]
test=df[~msk]
```

# Choosing the Model

```python
# Using Logistic Model to fit our data
# As logistic Function starts with a slow value and then increases so it may be the best fit for the values.
x=np.arange(-5.0 , 5.0 , 0.1)
y=1-4/(1+np.power(3,x-2))
plt.figure(figsize=(8,5))
plt.plot( x , y , color="green")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# Building up the Model

The formula for the logistic function is the following:

$$ \hat{Y} = \frac1{1+e^{\beta_1(X-\beta_2)}}$$

$\beta_1$: Controls the curve's steepness,

$\beta_2$: Slides the curve on the x-axis.

```python
def sigmoid(x , Beta_1 , Beta_2):
    return 1/(1+np.exp(-Beta_1*(x-Beta_2)))
```

```python
x_train=np.asanyarray(train[["Year"]])
y_train=np.asanyarray(train[["Value"]])
Beta_1=0.10
Beta_2=1990.0
y_hat=15000000000000*sigmoid(x_train , Beta_1 , Beta_2)
plt.plot(x_train , y_hat , color="red")
plt.plot(x_train , y_train , "bo")
plt.xlabel("Year")                                        
plt.ylabel("GDP")
plt.show()   
```

# Normalising Data

```python
x_train=x_train/max(x_train)
m,n=x_train.shape
y_train=y_train/max(y_train)   
```

# To Find Best Parameters , i.e. , Beta_1 amd Beta_2

We can use __curve_fit__ which uses non-linear least squares to fit our sigmoid function, to data. Optimal values for the parameters so that the sum of the squared residuals of sigmoid(xdata, *popt) - ydata is minimized.

popt are our optimized parameters.

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(sigmoid, x_train.reshape(m,) ,  y_train.reshape(m,))
#print the final parameters
print(" Beta_1 = %f, Beta_2 = %f" % (popt[0], popt[1]))
```

```python
print(popt)
```

```python
print(pcov)
```

```python
x = np.linspace(1960, 2015, 55)
x = x/max(x)
plt.figure(figsize=(8,5))
y = sigmoid(x, *popt)
plt.plot(x_train, y_train, 'ro', label='Data')
plt.plot(x,y, linewidth=3.0, label='Fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

# Performance on Train Set

```python
plt.figure(figsize=(8,5))
y_hat_train = sigmoid(x_train, *popt)
plt.plot(x_train, y_train, 'ro', label='Data')
plt.plot(x_train , y_hat_train , linewidth=3.0, label='Fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_hat_train - y_train)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_hat_train - y_train) ** 2))
```

# Performance on Test Set

```python
x_test=np.asanyarray(test[["Year"]])
y_test=np.asanyarray(test[["Value"]])
j,l=x_test.shape
y_test=y_test/max(y_test)   
x_test=x_test/max(x_test)
```

```python
popt1, pcov2= curve_fit(sigmoid, x_test.reshape(j,) ,  y_test.reshape(j,))
#print the final parameters
print(" Beta_1 = %f, Beta_2 = %f" % (popt1[0], popt1[1]))
```

```python
plt.figure(figsize=(8,5))
y_hat_test = sigmoid(x_test, *popt)
plt.plot(x_test, y_test, 'ro', label='Data')
plt.plot(x_test , y_hat_test , linewidth=3.0, label='Fit')
plt.legend(loc='best')
plt.ylabel('GDP')
plt.xlabel('Year')
plt.show()
```

```python
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_hat_test - y_test)))
print("Residual sum of squares (MSE): %.2f" % np.mean((y_hat_test - y_test) ** 2))
```
